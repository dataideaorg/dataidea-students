```python
# Data preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from dataidea.tabular import *
```

```python
#loading the dataset
customer_data=pd.read_csv('datasets/Mall_Customers.xls')

```

```python
# Getting to know the first five 5 rows.
customer_data.head()
```

```python
# Getting the know the number in rows in the dataset
customer_data.shape
```

```python
#Getting to know if there are any missing values in the dataset
customer_data.isnull().sum()
```

```python
#We choosing the annual income and spending score column

x=customer_data.iloc[:,[3,4]].values
```

```python
x
```

```python
#finding the number of clusters using the wcss(within clusters sum of squares) paramater

wcss=[]

for i in range(1,30):

    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
# second method of getting  number of clusters
range_=range(1,30)
inertia=[]
for n_clusters in range_:
    model = KMeans(n_clusters , random_state=0).fit(x)
    inertia.append(model.inertia_)
```

```python
#getting the elbow curve
plt.plot(range_, inertia)
plt.title('The Elbow curve')
plt.xlabel('Number of clusters')
plt.ylabel('inertia')
```

```python
#plotting the elbow curve to know the appropriate number of cluster
sns.set()
plt.plot(range(1,30), wcss)
plt.title('The elbow curve')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# Number of clusters=5

# Trainging the KMeans or fitting the clusters into the kmeans variable.

kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)

#Since they are 5 clusters we shall get the labels of all the 5 clusters and grouping all the data points according to labels

y = kmeans.fit_predict(x)
```

```python
y
```

```python
len(y)
```

```python
# Visualising the data .
plt.figure(figsize=(7,7))
plt.scatter(x[y==0, 0]   , x[y==0,1] , s=50 , color='blue' ,label='cluster 1')
plt.scatter(x[y==1, 0]   , x[y==1,1] , s=50 , color='yellow' ,label='cluster 2')
plt.scatter(x[y==2, 0]   , x[y==2,1] , s=50 , color='red' ,label='cluster 3')
plt.scatter(x[y==3, 0]   , x[y==3,1] , s=50 , color='green' ,label='cluster 4')
plt.scatter(x[y==4, 0]   , x[y==4,1] , s=50 , color='orange' ,label='cluster 5')
plt.title('Clustered customer groups')
plt.xlabel('The Annual income')
plt.ylabel('The customer spending score(1-100)')
# plotting the centroids
plt.scatter(kmeans.cluster_centers_[:,0]   ,     kmeans.cluster_centers_[:,1],  s=100,color='black',label='centroids')
plt.legend()
plt.savefig('.pdf')
plt.show()
```
