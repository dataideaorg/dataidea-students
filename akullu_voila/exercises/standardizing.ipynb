```python
# (x-mean)/standard_deviation
import numpy as np

def standardizer(values):
    mean_value = np.mean(values)
    standard_div = np.std(values)
    
    standardized = [(value-mean_value)/standard_div for value in values]
    return standardized    
```

```python
ages = [15,20,25,30,35,40,45,50]

# age = np.mean(ages)

standard_div = np.std(ages)

# print(age)
# print(standard_div)

standardizer(values=ages)
```
