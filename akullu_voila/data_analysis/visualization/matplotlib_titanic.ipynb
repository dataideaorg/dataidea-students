```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(filepath_or_buffer='../datasets/titanic3.csv')
```

```python
# plot a bar graph. 
data.pclass.value_counts()
```

```python
classes = data.pclass.value_counts().index
```

```python
values = data.pclass.value_counts().values
```

```python
plt.bar(x=classes, height=values)
```

```python
# histogram: Age distribution of passengers

plt.hist(x=data.age, bins=20)
plt.show()
```

```python
# Stacked Bar Chat: Survial status within each passenger class
survival_counts = data.groupby(by=['pclass', 'survived']).size().unstack()
survival_counts.plot(kind='bar', stacked=True)
```

```python
# Pie Chart: Proportion of male to fremale passengers
gender_counts = data.sex.value_counts()
plt.pie(x=gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.show()
```
